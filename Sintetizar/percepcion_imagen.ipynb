```python
from skimage import io, color

img = io.imread("img/mario.jpg")
io.imshow(img)
print(img) #imprime matriz de colores
```

```python
#print(img.shape) #escala de colores
imgg = io.imread("img/hongo.jpg")
n = color.rgb2gray(imgg)
io.imshow(n)
io.show()
print(imgg.shape)
```

```python
#Transformacion blanco y negro puro
white_black = color.rgb2gray(imgg)
rows, cols = white_black.shape

for i in range(rows):
	for j in range(cols):
		if(white_black[i,j] <= 0.5):
			white_black[i,j] = 0
		else:
			white_black[i,j] = 1
io.imshow(white_black)
io.show
```

```python
import skimage
vieja = io.imread("img/vieja.jpg")
#ruido gaussiano
noise = skimage.util.random_noise(vieja, mode="gaussian", var=0.0000001)

io.imshow(noise)
```
