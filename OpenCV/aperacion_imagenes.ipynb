## Sustraccion de imagenes

Sustraccion en escala de grises

```python
import cv2
img1 = cv2.imread('ojo.jpg',0)
img2 = cv2.imread('ojo2.jpg',0)
res = cv2.resize(img1,(400,300),interpolation = cv2.INTER_CUBIC)
res1 = cv2.resize(img2,(400,300),interpolation = cv2.INTER_CUBIC)
resultado = cv2.subtract(res,res1)
print('img1[0,0]= ',res[0,0])
print('img2[0,0]= ',res1[0,0])
print('resultado[0,0]= ',resultado[0,0])
cv2.imshow('resultado',resultado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img1 = cv2.imread('ojo.jpg',0)
img2 = cv2.imread('ojo2.jpg',0)
res = cv2.resize(img1,(400,300),interpolation = cv2.INTER_CUBIC)
res1 = cv2.resize(img2,(400,300),interpolation = cv2.INTER_CUBIC)

resultado2 = cv2.absdiff(res,res1)
print('img1[0,0]= ', img1[0,0])
print('img2[0,0]= ',img2[0,0])
print('resultado2[0,0]= ',resultado2[0,0])

cv2.imshow('resultado2',resultado2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture('video.mp4')
fgbg = cv2.bgsegm.createBackgroundSubtractorMOG()
while True:
    ret, frame = cap.read()
    if ret == False: break
    fgmask = fgbg.apply(frame)
    
    cv2.imshow('fgmask',fgmask)
    cv2.imshow('frame',frame)
    
    k = cv2.waitKey(30) & 0xFF
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
